# Importing Necessary Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, roc_auc_score
import os
```

# Reading Data

```python
data = pd.read_csv('datasets/Training_Data.csv')
```

# Pre Processing Data

```python
data = data.drop("Id", axis=1)
data.head()
```

```python
data = data.drop("CITY", axis=1)
```

## Encoding Categorical Data

```python
en = LabelEncoder()
catCols = ['Married/Single','House_Ownership','Car_Ownership','Profession','STATE']
for cols in catCols:
    data[cols] = en.fit_transform(data[cols])
```

```python
data.head()
```

```python
data.info()
```

# Splitting Data to X and Y

```python
Y = data["Risk_Flag"]
X = data.drop("Risk_Flag",axis = 1)
```

```python
X.head()
```

```python
Y.head()
```

```python
Y=pd.DataFrame(Y)
```

```python
Y.head()
```

```python
X.shape
```

```python
Y.shape
```

# Splitting Data as Test and Validation

```python
# Break off validation set from training data
X_train_full, X_valid_full, Y_train, Y_valid = train_test_split(X, Y, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
Y_train.head()
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train_full, Y_train)
predsdtc = dtc.predict(X_valid_full)
print("accuracy_score: " + str(accuracy_score(Y_valid, predsdtc)))
```

```python
print("f1_score: " + str(f1_score(Y_valid, predsdtc)))
```

```python
print("ROC AUC score: " + str(roc_auc_score(Y_valid, predsdtc)))
```
